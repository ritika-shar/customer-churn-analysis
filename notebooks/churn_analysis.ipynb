```python
sns.countplot(x='gender', hue='churn', data=df)
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='tenure', hue='churn', multiple='stack')
plt.title("Customer Tenure vs Churn")
plt.xlabel("Tenure (Months)")
plt.ylabel("Number of Customers")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x='contract_type', hue='churn', data=df)
plt.title("Churn by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Customer Count")
plt.show()
```

```python
plt.figure(figsize=(9,5))
sns.countplot(x='payment_method', hue='churn', data=df)
plt.title("Churn by Payment Method")
plt.xticks(rotation=30)
plt.xlabel("Payment Method")
plt.ylabel("Customer Count")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x='internet_service', hue='churn', data=df)
plt.title("Churn by Internet Service Type")
plt.show()
```

The analysis shows that customers using electronic check payment methods and fiber optic internet services are more likely to churn. Long-term contracts and automated payment methods significantly improve customer retention.
